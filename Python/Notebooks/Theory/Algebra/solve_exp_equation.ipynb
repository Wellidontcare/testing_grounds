```python
import sympy as sp
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
```

```python
x, u = sp.symbols('x, u', real=True)
```

```python
f = 3**x - 3**(0.5*(x+4)) + 20
```

```python
u_s = 3**(0.5*x)
f_sub = f.factor().subs(u_s, u).simplify()
display(f_sub)
```

```python
u1, u2 = sp.solve(f_sub, u)
display(u1)
```

```python
x1 = sp.solveset(sp.Eq(u1, u_s), x, domain=sp.S.Reals)
x2 = sp.solveset(sp.Eq(u2, u_s), x, domain=sp.S.Reals)
display(x1)
display(x2)
display(sp.Eq(u1, u_s))
```

```python
sp.solveset(sp.expand_power_base(f), x, domain=sp.S.Reals)
```

```python
display(sp.solveset(f.rewrite(sp.exp)))
```

```python
sp.nonlinsolve(f)
```
