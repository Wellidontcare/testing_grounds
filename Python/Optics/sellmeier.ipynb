```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sellmeier(Bs, Cs, lam):
    return np.sqrt(1 + sum((b*lam**2)/(lam**2 - c) for (b, c) in zip(Bs, Cs)))

def abbe(n_d, n_F, n_C):
    return (n_d - 1)/(n_F - n_C)


# BK7 Dispersion
Bs = [1.03961212, 0.231792344, 1.01046945]
Cs = [6.00069867e-3, 2.00179144e-2, 1.03560653e2]

lam = np.linspace(0.450, 0.8, 100)
plt.plot(lam, sellmeier(Bs, Cs, lam))



d_line = 0.5876
F_line = 0.4861
C_line = 0.6563

n_d = sellmeier(Bs, Cs, d_line)
n_F = sellmeier(Bs, Cs, F_line)
n_C = sellmeier(Bs, Cs, C_line)


V_abbe = abbe(n_d, n_F, n_C)

print(f"V_abbe = {V_abbe}")
```
